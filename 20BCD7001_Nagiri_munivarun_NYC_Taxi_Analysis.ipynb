# 4. Data visualization


## 4.1 Graph 1: Scatter Plot for Pickup and Dropoff Location


**Justification**: A scatter plot is suitable for visualizing the geographical distribution of pickup and dropoff locations using longitude and latitude coordinates. It helps identify patterns, clusters, and outliers in the spatial data.

**Insights/Trends**: The scatter plot provides insights into the concentration and spread of pickup and dropoff locations. Clusters of points may indicate popular areas, while outliers might represent less common locations. The color differentiation between pickup (green) and dropoff (blue) helps distinguish between the two.

**Observation**:
- The scatter plot illustrates the geographical distribution of both pickup and dropoff locations.
- Clusters of points are noticeable, indicating areas with high demand for rides. Outliers might represent less common pickup or dropoff locations.
- Color differentiation (green for pickup, blue for dropoff) helps distinguish between the two, revealing potential patterns in the customers ride.

**Conclusion**: The above observation could help us to tell which part of a city has more demand for taxis. So that no of drivers can be increased in that particular area.

```python
# Setting up the figure size
plt.figure(figsize=(12, 8))
# Scatter plot for pickup locations
plt.scatter(df['pickup_longitude'], df['pickup_latitude'], color='green', label='Pickup', alpha=0.5)
# Scatter plot for dropoff locations
plt.scatter(df['dropoff_longitude'], df['dropoff_latitude'], color='blue', label='Dropoff', alpha=0.5)
# Adding title and axis labels
plt.title('Pickup and Dropoff Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
# Adding legend
plt.legend()
# Displaying grid lines
plt.grid(True)
# Showing the plot
plt.show()
```

## 4.2 Graph 2: Bar Plot for Distribution of Passenger Counts

**Justification**: A bar plot is suitable for visualizing the distribution of categorical data (passenger counts). It provides a clear representation of the frequency of each category.

**Insights/Trends**: The bar plot shows the distribution of passenger counts, helping to identify common passenger group sizes. It reveals whether most trips have a single passenger, multiple passengers, or other variations.


**Observation**:

- The bar plot shows the distribution of the number of passengers in taxi rides.

- Most trips appear to have one or two passengers, as these counts have higher frequencies.

- Trips with a higher number of passengers are less common, as seen by the lower frequencies for larger counts.

**Conclusion**: The observations indicates that taxis like SUVs are not needed much. Instead the taxi company can invest on sedan and hatchback cars.

```python
# Count the occurrences of each passenger count value and sort by index
passenger_counts = df['passenger_count'].value_counts().sort_index()
# Set up the figure size
plt.figure(figsize=(8, 6))
# Create a bar plot
passenger_counts.plot(kind='bar', color='green', edgecolor='black')
# Add title and axis labels
plt.title('Distribution of Passenger Counts')
plt.xlabel('Number of Passengers')
plt.ylabel('Frequency')
# Rotate x-axis labels for better readability
plt.xticks(rotation=0)
# Display grid lines on the y-axis
plt.grid(axis='y')
# Show the plot
plt.show()
```

## 4.3 Graph 3: Line Plot for Average Trip Duration Over Time (Time-Series Plot)

**Justification**: A line plot is suitable for showing trends or patterns in a continuous dataset over time (time series data). It helps visualize how the average trip duration changes across different dates.

**Insights/Trends**: The line plot illustrates the trend in the average trip duration over time, allowing for the identification of patterns, peaks, or declines. This can provide insights into whether trip durations vary by specific dates or follow a particular trend.

**Observation**:

- The line plot depicts the average trip duration over different dates.
- Fluctuations in the line suggest variations in trip duration over time.
- Peaks or valleys may indicate days with longer or shorter average trip durations, which could tell us the busy days of a month.

**Conclusion**: The observations can help us tell when the drivers are more in demand and be ready with more drivers at that time.

```python
# Convert 'pickup_datetime' to datetime format
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
# Extract the date from 'pickup_datetime'
df['pickup_date'] = df['pickup_datetime'].dt.date
# Group by 'pickup_date' and calculate the mean of 'trip_duration' for each date
daily_trip_duration = df.groupby('pickup_date')['trip_duration'].mean()
# Set up the figure size
plt.figure(figsize=(12, 6))
# Create a line plot for average trip duration over time
daily_trip_duration.plot(color='purple', marker='o', linestyle='-', linewidth=2)
# Add title and axis labels
plt.title('Average Trip Duration Over Time')
plt.xlabel('Date')
plt.ylabel('Average Trip Duration (seconds)')
# Display grid lines
plt.grid(True)
# Show the plot
plt.show()
```
